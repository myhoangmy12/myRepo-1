```python
#BAR

import matplotlib.pyplot as plt
import numpy as np
height=[3,1,12,18,45]
bars=('A','B','C','D','E')
y=np.arange(len(bars))
plt.bar(bars,height)
#or plt.bar(y,height)
plt.title('A bar chart')
plt.xticks(y,bars,color='orange',rotation=45,fontweight='bold',fontsize=17,horizontalalignment='right')
#plt.tick_params(labelbottom=False)
plt.show();


```

```python
#Stacked bar
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font',weight='bold')
bar1=[12,38,1,8,22]
bar2=[28,7,16,9,14]
bar3=[25,3,23,25,17]

#height of bar1+bar2
bars=np.add(bar1,bar2).tolist()

#position of bar
r=[0,1,2,3,4]
names=['A','B','C','D','E']
barwidth=0.75

plt.bar(r,bar1,color='pink',edgecolor='white',width=barwidth)
plt.bar(r,bar2,color='red',bottom=bar1,edgecolor='white',width=barwidth)    
plt.bar(r,bar3,color='blue',bottom=bars,edgecolor='white',width=barwidth)    
plt.xticks(r,names,fontweight='bold')
plt.xlabel('Group')
plt.show()
```

```python
#Group bar chart
import numpy as np
import matplotlib.pyplot as plt
barwidth=0.25
#set height of bar
bar1=[12,30,1,8,22]
bar2=[28,16,6,5,10]
bar3=[29,3,24,25,17]

#set position
r1=np.arange(len(bar1))
r2=[x+barwidth for x in r1]
r3=[x+barwidth for x in r2]  

#plot
plt.bar(r1,bar1,color='pink',width=barwidth,label='var1')
plt.bar(r2,bar2,color='#f9bc86',width=barwidth,label='var2')
plt.bar(r3,bar3,color='blue',width=barwidth,label='var3')

#title x
plt.xlabel('Group',fontweight='bold')
plt.xticks([r+barwidth for r in range(len(bar1))],['A','B','C','D','E'])

plt.legend()
plt.show()
           
```

```python
#pie chart
import matplotlib.pyplot as plt
import seaborn as sns

names='A','B','C','D'
values=[12,11,3,30]
plt.figure(figsize=(8,8))
plt.pie(values,labels=names,labeldistance=1.15,
        wedgeprops={'linewidth':3,'edgecolor':'white'});
```

```python
#decor pie chart
import matplotlib.pyplot as plt
import seaborn as sns

names='A','B','C','D'
values=[12,11,3,30]
explode=[0.3,0.02,0.02,0.02]
plt.figure(figsize=(8,8))
colors=sns.color_palette('bright')
plt.pie(values,labels=names,colors=colors,
        autopct='%0.0f%%',
        explode=explode,
        shadow=True,
        startangle=90,
        textprops={'color':'green','fontsize':16},
        wedgeprops={'linewidth':6},
        frame='true',
        center=(0.1,0.1),
        rotatelabels='true')
plt.show()
```

```python
#SUB PLOT
import seaborn as sns
import matplotlib.pyplot as plt

df=sns.load_dataset('iris')
fig, axs = plt.subplots(2,2)  #2 hàng 2 cột 

sns.histplot(data=df,x="sepal_length",kde=True,color="skyblue",ax=axs[0,0])
sns.histplot(data=df,x="sepal_width",kde=True,color="pink",ax=axs[0,1])
sns.histplot(data=df,x="petal_length",kde=True,color="gold",ax=axs[1,0])
sns.histplot(data=df,x="petal_width",kde=True,color="olive",ax=axs[1,1])
```

```python
#subplots
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset('iris')
num=['sepal_length']
a=len(num)
b=3 
c=1
fig=plt.figure(figsize=(10,8))

#vẽ đồ thị 1
for i in num:
  plt.subplot(a,b,c)
  sns.distplot(df[i])
  c=c+1

#vẽ đồ thị 2
  plt.subplot(a,b,c)
  plt.boxplot(x=df[i])
  c=c+1

#vẽ đồ thị 3
  plt.subplot(a,b,c)
  sns.scatterplot(data=df,x=i,y="petal_width",hue="species")
  c=c+1
plt.tight_layout()
plt.show()
```
